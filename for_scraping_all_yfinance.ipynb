```python
import yfinance as yf
import csv
import pandas as pd
import os
import time


symbolfile = open("SYMBOLS.csv", "r")
data = list(csv.reader(symbolfile, delimiter=","))
symbolfile.close()


dataList = []
for sublist in data:
    for item in sublist:
        dataList.append(item)

dataList.remove("Symbol")
print(dataList)
```

```python
inc = 0

for i in dataList:
    
    tickerSym = yf.Ticker(f"{dataList[inc]}")
    pathE = f'C:\\Users\\joshu\\Desktop\\Homeworks\\New folder\\Earnings\\{dataList[inc]}-earnings.csv'
    pathH = f'C:\\Users\\joshu\\Desktop\\Homeworks\\New folder\\Historical\\{dataList[inc]}-history.csv'
    
    
    try:
        historySym = tickerSym.history(period="max", interval="1d")
        earningsSym = tickerSym.earnings_dates
        earningsSym.to_csv(pathE, index=True) 
        historySym.to_csv(pathH, index=True)
        print(f"The current ticker is {tickerSym}")
        inc = inc + 1
        time.sleep(3)
        
    
    except:
        print("An exception occurred")
        inc = inc + 1
    
```

```python
inc = 0
unscrapedSymbolList = []

for i in dataList:
    
    #try except, because some {dataList[inc]} values were passed over in scraping
    
    try:
    
        pathE = f'C:\\Users\\joshu\\Desktop\\Homeworks\\New folder\\Earnings\\{dataList[inc]}-earnings.csv'
        pathH = f'C:\\Users\\joshu\\Desktop\\Homeworks\\New folder\\Historical\\{dataList[inc]}-history.csv'
    
    #Open earningsFile in order to clean dates in preparation for a merge
        earningsFile = open(pathE, "r")
        earningsData = list(csv.reader(earningsFile, delimiter=","))
        earningsFile.close()
    #to pd
        earningsDF = pd.DataFrame(data=earningsData)
    #clean dates from earningsDF using df['Date'] = pd.to_datetime(df['DateTime']).dt.date
        earningsDF.columns = earningsDF.iloc[0]
        earningsDF = earningsDF.drop(labels=[0],axis=0)
        earningsDF['Dates'] = earningsDF['Earnings Date'].str.split(' ').str[0]
        new_cols = ["Dates","EPS Estimate","Reported EPS","Surprise(%)"]
        earningsDF = earningsDF.reindex(columns=new_cols)
        earningsDF['Dates'] = pd.to_datetime(earningsDF["Dates"], format="%Y/%m/%d")
            
    
    #Open historyFile in order to clean dates in preparation for a merge
        historyFile = open(pathH, "r")
        historyData = list(csv.reader(historyFile, delimiter=","))
        historyFile.close()
    
    #to pd
        historyDF = pd.DataFrame(data=historyData)
    
    #clean dates from historyDF
        historyDF.columns = historyDF.iloc[0]
        historyDF = historyDF.drop(labels=[0],axis=0)
        historyDF['Dates'] = historyDF['Date'].str.split(' ').str[0]
        new_cols = ["Dates","Open","High","Low", "Close", "Volume", "Dividends", "Stock Splits"]
        historyDF = historyDF.reindex(columns=new_cols)
        historyDF['Dates'] = pd.to_datetime(historyDF["Dates"], format="%Y/%m/%d")
        
    #merge earnCleanDF and histCleanDF on ['Dates'] into a new DF, called mergedDF
        mergedEHDF = pd.merge(historyDF, earningsDF, on="Dates", how="left")
    
    #add a column of the dataList[inc]
        mergedEHDF.insert(0, "Symbol", f"{dataList[inc]}", allow_duplicates=True)    
    
    #then mergedDF.to_csv(pathMerged, index=True)
        pathEHMerged = f'C:\\Users\\joshu\\Desktop\\Homeworks\\New folder\\Merged\\{dataList[inc]}-merged.csv'
        mergedEHDF.to_csv(pathEHMerged, index=True)
    
    
    
    #add to inc and except out
        inc = inc + 1
    except:
    
    #as except only triggers when a Ticker at {dataList[inc]} cannot be scraped, save this Symbol
    #in order to create a new version of the ticker list of unscraped symbols. These are saved to a 
    #csv in order to try this process again on symbols that were unscraped for whatever reason
        
        unscrapedSymbolList.append(f"{dataList[inc]}")
        inc = inc + 1
        
unscrapedSymbolDF = pd.DataFrame(unscrapedSymbolList)
pathUnscraped = f'C:\\Users\\joshu\\Desktop\\Homeworks\\New folder\\Unscraped\\unscrapedSymbols.csv'
unscrapedSymbolDF.to_csv(pathUnscraped, index=True)
```
