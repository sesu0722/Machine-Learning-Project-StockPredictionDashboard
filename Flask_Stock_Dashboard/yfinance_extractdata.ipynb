```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import json
import pymongo
from pymongo import MongoClient
import datetime
from datetime import datetime
```

```python
import yfinance as yf

aapl= yf.Ticker("aapl")
aapl
```

```python
aapl_histl=yf.download("AAPL",start="2002-01-01", end="2022-11-15", interval="1d")

aapl_histl=aapl_histl.dropna()

aapl_histl['Ticker']= 'AAPL'

aapl_histl
```

```python
%matplotlib inline
plt.style.use("fivethirtyeight")

plt.figure(figsize=(14,7))
aapl_histl['Close'].plot()

plt.title('Apple Price data', fontsize=14)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Price', fontsize=12)

plt.show()
```

```python

price = aapl_histl["Close"]

fig, ax = plt.subplots(figsize=(14, 6))

half_year_locator = mdates.MonthLocator(interval=10)
ax.xaxis.set_major_locator(half_year_locator) # Locator for major axis only.

ax.plot(price);

fig.autofmt_xdate()
```

```python
data = yf.download("AMZN AAPL", start="2002-01-01", end="2022-10-30",group_by='Date').reset_index()

data['Date']=pd.to_datetime(data['Date'], format='%YYYY-%b-%d')

data
```

```python
data.dtypes
```

```python
#save json file
data.to_json('amznAapl.json')
```

```python

# Making Connection
myclient = MongoClient("mongodb://localhost:27017/")
  
# database
db = myclient["stockDB"]
  
# Created or Switched to collection

col1=db["amznAapl"]

# Loading or Opening the json file
with open('amznAapl.json') as file:
    file_data = json.load(file)

# Inserting the loaded data in the Collection
# if JSON contains data more than one entry
# insert_many is used else insert_one is used
if isinstance(file_data, list):
    col1.insert_many(file_data)
else:
    col1.insert_one(file_data)
```
